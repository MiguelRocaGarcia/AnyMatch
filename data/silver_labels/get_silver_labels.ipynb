```python
import pandas as pd
```

```python
df_anymatch_preds = pd.read_csv('../alliance/anymatch_predictions_full_v3.csv')
df_det_rules_pred = pd.read_csv('../alliance/deterministic_rules_predictions_full_v1.csv')
```

```python
df = df_det_rules_pred.merge(df_anymatch_preds, how='inner', on=['PATID_A', 'PATID_B'])
```

```python
def make_bool(m):
    return True if m == 'match' else False
df['silver_label'] = df.apply(lambda row: make_bool(row['rule_pred']) or bool(row['pred']), axis=1)
```

```python
df = df[['PATID_A', 'PATID_B','silver_label']]
```
